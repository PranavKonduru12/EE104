```python
from time import sleep
from pynq.overlays.base import BaseOverlay
base = BaseOverlay("base.bit")
```

```python
led0 = base.leds[0] #Corresponds to LED LD0
led1 = base.leds[1] #Corresponds to LED LD1
led2 = base.leds[2] #Corresponds to LED LD2
led3 = base.leds[3] #Corresponds to LED LD3
```

```python
rgb0 = base.rgbleds[4] #RGB LED0
rgb1 = base.rgbleds[5] #RGB LED1
```

```python
sw0 = base.switches[0] #Corresponds to SW0
sw1 = base.switches[1] #Corresponds to SW1
```

```python
b0 = base.buttons[0] #Corresponds to Button0
b1 = base.buttons[1] #Corresponds to Button1
b2 = base.buttons[2] #Corresponds to Button2
b3 = base.buttons[3] #Corresponds to Button3
```

```python
while(sw0.read() == True):  # All the code below while(True) runs forever
    if (b0.read() == True):   # Reads b0 and check if it pushed
        led0.on()              # IF b0 is ON --> Turn on LED0
    if (b1.read() == True):   # Reads b1 and check if it pushed
        led1.on()              # IF b1 is ON --> Turn on LED1
    if (b2.read() == True):   # Reads b2 and check if it pushed
        rgb0.write(7) #--> Turn on RGB LED1
        rgb1.write(5) #--> Turn on RGB LED1
    if (b3.read() == True):   # Reads b3 and check if it pushed
        rgb0.write(3) #--> Turn off RGB LED1
        rgb1.write(4) #--> Turn off RGB LED1
    if (sw1.read() == True):   # Reads sw1 and check if it toggled then turn off all LEDs
        led0.off()            
        led1.off()            
        rgb0.off() #--> Turn off RGB LED0
        rgb1.off() #--> Turn off RGB LED1
        

```

```python
while(True):  # All the code below while(True) runs forever
    if (sw0.read() == True):   # Reads SW0 and check if it toggled
        led0.on()              # IF SW0 is ON --> Turn on LED0
        led1.on()              # IF SW0 is ON --> Turn on LED1
    else:
        led0.off()             # ELSE Turn off LED0
        led1.off()             # ELSE Turn off LED1
    
    if (sw1.read() == True):   # Reads SW1 and check if it toggled
        led2.on()              # IF SW1 is ON --> Turn on LED2
        led3.on()              # IF SW1 is ON --> Turn on LED3
    else:
        led2.off()             # ELSE Turn off LED2
        led3.off()             # ELSE Turn off LED3
```
