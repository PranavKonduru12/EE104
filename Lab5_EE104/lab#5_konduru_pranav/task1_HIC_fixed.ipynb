```python
import math as m
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as integ
```

### With Airbag 
Plot a(t):
$$
  a(t)=\frac{22000}{(t-74)^2+500}
$$

```python
# Defined a(t) here
def a1(x):
    #x = [x/2 for x in range(300)]
    y = 22000 / ((x - 74)**2 + 500)
    return y
```

```python
# Plotted the function here
x = [x/2 for x in range(300)]
y = [a1(v) for v in x]
plt.plot(x, y) 
```

```python
integ.quad(a1, 0, 50) 
```

```python
def hic(t, d, a):
    #x = [x/2 for x in range(300)]
    #y = [a1(v) for v in x]
    return d * ((1/d) * integ.quad(a, t, t+d)[0]) ** 2.5
```

```python
hic(0, 50, a1)
```

```python
yi = [hic(v, 50, a1) for v in x]
plt.plot(x, yi)
```

### Without Airbag
Plot a(t):
$$
  a(t)=\frac{16400}{(t-68)^2+400}+\frac{1480}{(t-93)^2+18}
$$

```python
# Defined a(t) here
def a2(x):
    #x = [x/2 for x in range(300)]
    y = (16400 / ((x - 68)**2 + 400)) + (1480 / ((x - 93)**2 + 18))
    return y
```

```python
y = [a2(v) for v in x]
plt.plot(x, y) 
```

```python
integ.quad(a2, 0, 50) 
```

```python
hic(0, 50, a2)
```

```python
yj = [hic(v, 50, a2) for v in x]
plt.plot(x, yj)
```
